```python
def word_count(str):
    counts = dict()
    words = str.split()

    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1

    return counts

value_counts= word_count(result)

```

```python
stopwords = ['the', 'of', 'and', 'a', 'to', 'in','is', 'with', 'height', 'width', 'alt', 
             'for', 'The', 'his', 'by', 'on', 'this', 'as', 'Radio', 'Drama', 'week',
             'This', 'that', 'from', 'an', 'we', 'Episode', 'you', 
             'I', 'it', 'he', 'be', 'who', 'out', 'at', 'into', 'two', 'all', 'up', 'their', 'me',
            'have', 'our', 'us', 'are', "more", "It's", 'she', 'her', 'was', 'so', 'not', "We", 'one',
            'but', 'they', 'as', 'him', 'new', 'talk', 'show', 'podcast', 'radio', 'drama', 'episode',
            'production']
querywords = corpus.split()

resultwords  = [word for word in querywords if word.lower() not in stopwords]
result = ' '.join(resultwords)

print(result)
```

```python
mask = np.array(Image.open("100420169-radio-tower-mast-with-radio-waves-for-broadcast-transmission-line-art-vector-icon-for-apps-and-websi.jpg"))
mask
```

```python
def transform_format(val):
    if val == 0:
        return 255
    else:
        return val
    
    
```

```python
transformed_wine_mask = np.ndarray((mask.shape[0],mask.shape[1]), np.int32)

for i in range(len(mask)):
    transformed_wine_mask[i] = list(map(transform_format, mask[i]))

transformed_wine_mask
```

```python
wc = WordCloud(background_color="white", max_words=3000, mask= transformed_wine_mask,  contour_width=3, contour_color='steelblue')

# Generate a wordcloud
wc.generate(result)

# store to file
#wc.to_file("RDR3.png")

# show
plt.figure(figsize=[30,30])
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import os

from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

wc = WordCloud(background_color="white", max_words=2000, mask=alice_coloring,
                max_font_size=40, random_state=42)
# generate word cloud
wc.generate(result)

# create coloring from image
image_colors = ImageColorGenerator(alice_coloring)

# show
fig, axes = plt.subplots(1, 3)
axes[0].imshow(wc, interpolation="bilinear")
# recolor wordcloud and show
# we could also give color_func=image_colors directly in the constructor
axes[1].imshow(wc.recolor(color_func=image_colors), interpolation="bilinear")
axes[2].imshow(alice_coloring, cmap=plt.cm.gray, interpolation="bilinear")
for ax in axes:
    ax.set_axis_off()
plt.show()
```
